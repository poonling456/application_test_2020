```python
#Import packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
from statsmodels.stats import outliers_influence

#Check workinig directory
#os.getcwd()

#Function for VIF calculation
def calc_vif(X):
    # Calculating VIF
    vif = pd.DataFrame()
    vif["variables"] = X.columns
    vif["VIF"] = [outliers_influence.variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    return(vif)
```

```python
#import data
df = pd.read_csv(r'DATA_GPA modeling.csv',index_col=[0])

#Drop NA values
df_dropna = df.dropna()

#verify no null values
df_dropna.isnull().sum().sum()
```

```python
#Settnig X Y Variables
from sklearn import linear_model
import statsmodels.api as sm
X = df_dropna[['meals', 'ell', 'yr_rnd',
       'mobility', 'acs_k3', 'acs_46', 'not_hsg', 'hsg', 'some_col', 'avg_ed',
       'full', 'emer', 'enroll']]
Y_00=df_dropna['api00']
Y_avg_results = (df_dropna['api00']+df_dropna['api99'])/2
```

```python
###############################
# Step 0 :Using OLS modelling #
###############################

#statsmodels regression for Y_avg_results data
#using AVG results becuase
# 1) high correlation between 00 99
# 2) growth mostly positive, might be systematic change in whole country
# 3) growth data uncorrelated with all varaibles
# 4) average results can express the academic performances, as its hardly be greatly improved within a year interval.
X = sm.add_constant(X)
model = sm.OLS(Y_avg_results, X).fit()
print(model.summary())

plt.rc('figure', figsize=(12, 7))
plt.text(0.01, 0.05, str(model.summary()), {'fontsize': 10}, fontproperties = 'monospace') # approach improved by OP -> monospace!
plt.axis('off')
plt.tight_layout()
plt.savefig(r'OLS full model(w constant).png')

#Conclusion: 
#kick out variables with large p values one by one= constant, yr_rnd, mobility, acs_k3, full, enroll
#Limitation: 
#Tackle to limitation:
```

```python
#################################
# Step 0.1 :Using OLS modelling #
#################################

#OLS skippnig large p values
X_p = df_dropna[['meals', 'ell', 'acs_46', 'not_hsg', 'hsg', 'some_col', 'avg_ed','emer']]
model_p = sm.OLS(Y_avg_results, X_p).fit()
print(model_p.summary())

plt.rc('figure', figsize=(12, 7))
plt.text(0.01, 0.05, str(model_p.summary()), {'fontsize': 10}, fontproperties = 'monospace') 
plt.axis('off')
plt.tight_layout()
plt.savefig('OLS model_p.png')
```

```python
###############################
#Step 1 :Test autocorrelation #
###############################

#testing autocorrelation between api00 api99 (high correlation)
from statsmodels.stats.stattools import durbin_watson
model_autocorr = sm.OLS(df_dropna['api00'], df_dropna['api99']).fit()
print('DW-test statistics:' + str(durbin_watson(model_autocorr.resid)))

#Conclusion: 
#This case here only have two time points, horizontally across different schools, autocorrelation test is ill-defined.
#We shall skip the following conclusions and continue using OLS
#1.736 > 1.664 = autocorrelation
#autocorrelation is marginally problematic
#
#autocorrelation can be also tested for cross sectional data, however fundamentals of the data are not provided and the test will be ill-defined.
```

```python

#######################################
# Step 2.1 :Testing multicollinearity #
#######################################

#Testing multicollinearity with VIF
#Rule of thumb: VIF>10

#X_p = df_dropna[['meals', 'ell', 'acs_46', 'not_hsg', 'hsg', 'some_col', 'avg_ed','emer']]

#VIF for streamed model
print(calc_vif(X_p))
#Conclusion: 
# 1) Collinearity for "acs46"

#VIF for streamed model，skiped "acs46"
print(calc_vif(X_p[['meals', 'ell', 'not_hsg', 'hsg', 'some_col', 'avg_ed','emer']]))
#Conclusion:
# 1) Collinearity for "meals"

#VIF for streamed model，skiped "acs46", "meals"
print(calc_vif(X_p[['ell', 'not_hsg', 'hsg', 'some_col', 'avg_ed','emer']]))
#COnlusion:
# 1) Coolinearity for "meals" with "not hsg"
# 2) no more VIF > 10, most collinearity problem diagnosed

X_VIF = X_p[['ell', 'not_hsg', 'hsg', 'some_col', 'avg_ed','emer']]
```

```python
#######################################
# Step 2.2 :PCA for multicollinearity #
#######################################
#PCA for collinearity, after trimming for high p-value model


from sklearn.decomposition import PCA

#check correlation for X_p
corrmat=X_p.corr()
sns.heatmap(corrmat, annot=True)
plt.show()

#apply PCA with 3 components (= explained 93% of variance)
pca = PCA(n_components=3)
principalComponents = pca.fit_transform(corrmat)
print(pca.explained_variance_ratio_.sum())
#Conclusions:
# 1） creating new variables that are totally independent
# 2) taking 3 PCs, already explaining 93% of vairance

#dataframe for PC
feature_pca = ['PC'+str(i+1) for i in range(pca.n_components_)]
PCs = pd.DataFrame(principalComponents,columns = feature_pca)
PCs.index = X_p.columns
print(PCs)

#heatmap for corr of PCs --> corr = 0, perfectly uncorrelated
PCs.corr().style.background_gradient(cmap='coolwarm').set_precision(2)
plt.show()
```

```python
#######################################
# Step 2.2 :PCA for multicollinearity #
#######################################

#Model for PCA, each entry has 3Xs = 3 PCs = 3 Variables
X_PCA = pca.transform(X_p)
model_pca = sm.OLS(Y_avg_results, X_PCA).fit()

print(model_pca.summary())

plt.rc('figure', figsize=(12, 7))
plt.text(0.01, 0.05, str(model_pca.summary()), {'fontsize': 10}, fontproperties = 'monospace') # approach improved by OP -> monospace!
plt.axis('off')
plt.tight_layout()
plt.savefig(r'OLS PCA model(wo constant).png')
```

```python
#######################################
# Step 2.3 :PCA for multicollinearity #
#######################################

#Check PVA dimensionality reduction and compare results between LR model and LR with 3PC
#by evaluating a cross validation score using cross_val_score

regression = linear_model.LinearRegression()

import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

from sklearn.model_selection import cross_val_score

#Cross Validations scores for all 
X_score = cross_val_score(regression, X, Y_avg_results).mean()
X_P_score = cross_val_score(regression, X_p, Y_avg_results).mean()
X_VIF_score = cross_val_score(regression, X_VIF, Y_avg_results).mean()
X_PCA = pca.transform(X_p)
X_PCA_score = cross_val_score(regression, X_PCA, Y_avg_results).mean()

#Cross_val_score_table = pd.DataFrame(dict(X_score=X_score,X_P_score=X_P_score,X_VIF_score=X_VIF_score,X_PCA_score=X_PCA_score))
Cross_val_score_table=pd.DataFrame(pd.Series([X_score,X_P_score,X_VIF_score,X_PCA_score]).rename('Cross Validation Score'))
Cross_val_score_table.index = ['X_score','X_P_score','X_VIF_score','X_PCA_score']
print(Cross_val_score_table)
```

```python
######################################
# Step 3.1 testing Heteroskedasticity#
######################################

from statsmodels.stats.diagnostic import het_breuschpagan
from statsmodels.stats.diagnostic import het_white
from statsmodels.formula.api import ols
from scipy import stats

#Method 1: QQ plot for X_p model
stats.probplot(model_p.resid, dist="norm", plot= plt)
plt.show()
plt.savefig("model_p_Resid_qqplot.png")
#Conclusion:
# 1) shows heteroskedasticity


#Method xxxx: Residual plots for all other X vairables
#for col in df_dropna.columns:
#    plt.figure()
#    plt.scatter(df_dropna[col],model.resid)
#    plt.title(col)
#    plt.show()
#Conclusion:
# 1) shows heteroskedasticity

    
#Method 2: Breusch-Pagan test
bp_test = het_breuschpagan(model.resid, model.model.exog)
labels = ['LM Statistic', 'LM-Test p-value', 'F-Statistic', 'F-Test p-value']
bp_test = pd.DataFrame(pd.Series(bp_test).rename('bp_test'))
bp_test.index = labels
print(bp_test)
#Conclusion: both p-value are <0.05 --> model is with Heteroskedastic
#Conclusoin: --> non consistent variance over different perioid
#WHy us BP not Whte:
# Make better us of data (more power) (white test formulas costs a lot of data)
# more computationally efficient
```

```python
######################################
# Step 3.2 bye bye Heteroskedasticity#
######################################

# Handle hetero (non-constant varaince) by fitting White's

#OLS on X_p by fitting White's
model_White = sm.OLS(Y_avg_results, X_p).fit(cov_type='HC1',use_t=None)
print(model_White.summary())

plt.rc('figure', figsize=(12, 7))
plt.text(0.01, 0.05, str(model_White.summary()), {'fontsize': 10}, fontproperties = 'monospace') 
plt.axis('off')
plt.tight_layout()
#plt.savefig('OLS model_p (w white robust estimator.png')
```
